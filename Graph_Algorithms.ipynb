```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
```

```python
G = nx.karate_club_graph()
```

```python
plt.figure(figsize=(8, 8))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title("Zachary's Karate Club Graph")
plt.show()
```

```python
start_node = 0  #   
dfs_order = list(nx.dfs_preorder_nodes(G, source=start_node))
bfs_order = list(nx.bfs_tree(G, source=start_node))
print("DFS Order:", dfs_order)
print("BFS Order:", bfs_order)
```

```python
shortest_path = nx.shortest_path(G, source=0, target=33)
print("Shortest path from 0 to 33:", shortest_path)
```

```python
cycles = list(nx.cycle_basis(G))
print("Cycles:", cycles)
```

```python
adj_matrix = nx.adjacency_matrix(G).toarray()
print("Adjacency Matrix:\n", adj_matrix)
```

```python
values, vectors = eigh(adj_matrix)
print("Eigenvalues:\n", values)
print("Eigenvectors:\n", vectors)
```
